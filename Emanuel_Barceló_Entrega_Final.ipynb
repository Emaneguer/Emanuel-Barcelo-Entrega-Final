```python
# Datos de preferencias en las diferentes ramas del Dibujo Realista
data_str = """Cantidad de personas,Edades,Interes
75 personas,7 a 65 a√±os,Dibujo de paisajes
96 personas,15 a 45 a√±os,Dibujo del rostro
200 personas,10 a 35 a√±os,Dibujo de la figura humana
80 personas,12 a 35 a√±os,Dibujo de animales
20 personas,45 a 60 a√±os,Dibujo de bodeg√≥n
"""

# Crear un objeto StringIO para simular un archivo CSV
csv_file = StringIO(data_str)

# Leer el CSV y almacenar los datos en una lista de diccionarios
reader = csv.DictReader(csv_file)
data = list(reader)

# Mostrar los datos
print(data)
```

```python
prompt = "Quiero que tomando en cuenta los datos del contexto extraigas entidades y relaciones entre los datos con el objetivo de crear un gr√°fico {'Cantidad': '75 personas', 'Rango etario': '7 a 65 a√±os', 'Interes': 'Dibujo de paisajes'}, {'Cantidad': '96 personas', 'Rango etario': '10 a 35 a√±os' 'Interes': 'Dibujo del rostro' } solo tomalo como ejemplo"

prepared_prompt = f"Contexto:{data}\n {prompt}"
print(prepared_prompt)
```

```python
# Make the request
conversation = [
        {"role": "system", "content": "Eres un asistente util, tu objetivo principal es extraer todas las relaciones entre las entidades y sus datos,  retorna un array de objetos"},
        {"role": "user", "content": prepared_prompt},
    ]

# Make the request
response = openai.ChatCompletion.create(
    model= 'gpt-3.5-turbo',
    max_tokens=800,
    messages=conversation
    )

message = response.choices[0]['message']
print("{}: {}".format(message['role'], message['content']))
```

```python
import networkx as nx

G = nx.MultiDiGraph()

data = eval(message['content'])
print(f"data {data}")
#data = eval(str(data))

for d in data:
         G.add_edge(d['Cantidad'], d['Interes'],**d)
```

```python
pos = nx.spring_layout(G, k=0.8)  # layout spring for the nodes position
nx.draw(G, pos, with_labels=True, node_color='pink', node_size=2000, font_size=8, edge_color='blue', width=1, alpha=0.8)

# Etiquetas de los bordes
edge_labels = {(n1, n2): data.get('Rango etario') for n1, n2, data in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)
```
